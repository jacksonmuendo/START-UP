```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sample_data = pd.read_csv('iris.csv')
```

```python
ishmael = sample_data[sample_data.species == 'setosa']
#print(ishnael)
```

```python
jackson = sample_data[sample_data.species == 'virginica']
```

```python
lilian = sample_data[sample_data.species == 'versicolor']
```

```python
plt.scatter(lilian.sepal_length, lilian.sepal_width)
plt.scatter(jackson.sepal_length, jackson.sepal_width)
plt.scatter(ishmael.sepal_length, ishmael.sepal_width)
plt.title('iris chart')
plt.legend(['lilian', 'jackson', 'ishmael'])
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

```python
plt.scatter(lilian.petal_length, lilian.petal_width)
plt.scatter(jackson.petal_length, jackson.petal_width)
plt.scatter(ishmael.petal_length, ishmael.petal_width)
plt.legend(['lilian', 'jackson', 'ishmael'])
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.title('iris_sampling')
plt.show()
```
